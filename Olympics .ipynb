```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib%inline
```

```python
#Importing datasets 

athletes=pd.read_csv(r'C:\Users\yashr\Downloads\archive (15)/athlete_events.csv')
region=pd.read_csv(r'C:\Users\yashr\Downloads\archive (15)/noc_regions.csv')
```

```python
# Gold Medals By Region

goldmedal['Region'].value_counts().reset_index(name='Medals').head()
```

```python
# plot for Gold Medals By Region

Top_goldmedal=goldmedal['Region'].value_counts().reset_index(name='Medals').head()
g=sns.catplot(x="index", y="Medals", data=Top_goldmedal, height=5, kind='bar', palette='rocket')
plt.title('Gold medals per country')
g.despine(left=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('No. of Medals')
```
